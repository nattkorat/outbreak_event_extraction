```python
import json
from time import sleep
from random import randint
from utils import prompt, response_parser
from llms import gemini2_5
```

```python
# load data
with open('data/20250618_TH_samples_21.json', 'r') as f:
    data = json.load(f)

print(data[0])
```

```python
data_with_responses = []
l = len(data)
i = 0
while i < l:
    print(f'Processing {i + 1}/{l}...')
    item = data[i]
    try:
        response = gemini2_5.chat_with_gemini2_5(
            text=prompt.event_extraction_prompt_zero_shot(
                item['content'],
            )
        )
        response = response_parser.json_string_response_parser(response)
        item['response'] = response
        data_with_responses.append(item)
        i += 1
        print(f'Item {i} processed successfully.')
        sleep(randint(1,3))  # to avoid rate limiting
        
    except Exception as e:
        print(f'Error processing item {i + 1}: {e}')
        # if an error occurs, wait and retry
        print('Retrying after a short wait (65 seconds)...')
        sleep(65)  # wait before retrying
        
```

```python
data_with_responses[5]
```

```python
with open('data/20250618_TH_samples_21_with_responses_zero_shot.json', 'w') as f:
    json.dump(data_with_responses, f, indent=4, ensure_ascii=False)
```
