```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/29k_cleaned_thai_20250508_v2_token_count.csv')
data.head()
```

```python
sample_21 = data.sample(21, random_state=168)
sample_21.head()
```

```python
from matplotlib import rcParams

# Set font to a Thai-supported font (e.g., 'Tahoma' or 'Noto Sans Thai')
rcParams['font.family'] = 'Tahoma'

sample_21['site_name'].value_counts().plot(kind='bar', title='Sampled Site Names')
plt.xticks(rotation=0)
plt.xlabel('Count')
plt.show()
```

```python
data = sample_21.loc[:, ['title', 'site_name', 'content', 'token_count',]]
data
```

```python
import json

# Save data to JSON format (easier for other processes)
with open('data/20250618_TH_samples_21.json', 'w', encoding='utf-8') as f:
    json.dump(data.to_dict(orient='records'), f, ensure_ascii=False, indent=2)
```
