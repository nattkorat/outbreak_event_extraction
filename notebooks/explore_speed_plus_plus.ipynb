```python
import json 

# read jsonl file
def read_jsonl(file_path, limited_lines = None):
    data = []
    i = 0
    with open(file_path, 'r') as file:
        for line in file:
            data.append(json.loads(line))
            i += 1
            if limited_lines is not None and i >= limited_lines:
                break
            
    return data
```

```python
data = read_jsonl('data/data-speed++/en-covid-dev.jsonl', limited_lines=15)
data
```

```python
test = []
for d in data:
    temp = {}
    temp['text'] = d['sentence']
    temp['events'] = {}
    for e in d['event_mentions']:
        e_type = e['event_type']
        if e_type not in temp['events']:
            temp['events'][e_type] = []
        temp['events'][e_type].append({
            'trigger': e['trigger']['text'],
            'arguments': {arg['role']: arg['text'] for arg in e['arguments']}
        })
    test.append(temp)
test
```

```python
templates, test = test[:5], test[5:]
```

```python
# save as json for testing models
with open('speedpp_templates.json', 'w') as f:
    json.dump(templates, f, ensure_ascii=False, indent=2)
# save as json for testing models
with open('speedpp_test.json', 'w') as f:
    json.dump(test, f, ensure_ascii=False, indent=2)
```

## Add ID to Test samples

```python
import json
from uuid import uuid4

# read json file
with open('speedpp_test.json', 'r') as f:
    samples = json.load(f)

# add id to each samples
for i, sample in enumerate(samples):
    sample['id'] = str(uuid4())
```
